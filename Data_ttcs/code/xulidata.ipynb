```python
ratings=[]
for i in df['ratings']:
    if len(i)>0:
        str=i[-1]
        str=str.replace("\n", "")
        str=str.replace("\xa0", "")
        str=str.replace(" đánh giá", "")
        ratings.append(str)
    else:
        ratings.append(0)
ratings
df['ratings']=ratings
```

```python
size=[]
for i in df['size']:
    
    size.append(i.replace(" G",""))
df['size']=size
df['size']=df['size'].astype(np.float64)
df.head()
```

```python
price=[]
for i in df['price']:
    if i!=None:
        price.append(i.replace("\xa0₫",""))
    else:
        price.append(None)
df['price']=price

# df['price']=df['price'].apply(np.int64)
df.head()
```

```python
price=[]
for i in df['price']:
    if i!=None:
        price.append(i.replace(".",""))
    else:
        price.append(0)
df['price']=price

df['price']=df['price'].apply(np.int64)
df['price']
```

```python
price_selloff=[]
for i in df['price_selloff']:
    if i!=None:
        x=i.replace("\xa0₫","")
        price_selloff.append(x.replace(".",""))
    else:
        price_selloff.append(0)
df['price_selloff']=price_selloff

df['price_selloff']=df['price_selloff'].apply(np.int64)
df['price_selloff']
```

```python
import plotly.graph_objects as go
labels = ['iphone','xiaomi','nokia','oppo','samsung']
values=[0,0,0,0,0]
for i in df["brand"]:
    if i==labels[0]:
        values[0]+=1
    if i==labels[1]:
        values[1]+=1
    if i==labels[2]:
        values[2]+=1
    if i==labels[3]:
        values[3]+=1
    if i==labels[4]:
        values[4]+=1    
print(values)
colors = ['gold', 'mediumturquoise', 'darkorange', 'lightgreen','darkblue']

fig = go.Figure(data=[go.Pie(labels=labels,
                             values=values)])
fig.update_traces(hoverinfo='label+percent', textfont_size=20,
                  marker=dict(colors=colors, line=dict(color='#000000', width=2)))
fig.show()
```
